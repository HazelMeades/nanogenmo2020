```python
import random
import markovify
```

```python
novel = ''

# Get raw text as string.
with open("nanowrimo/combined-finished-nanos-nochapters.txt") as f:
    text = f.read()
    
# Build the model.
text_model = markovify.Text(text)

# Print five randomly-generated sentences
for i in range(3600):
    
    novel += str(text_model.make_sentence()) + ' '
    
    r = random.randint(0,100)
    
    if (r<6):
        novel += '\n\n'
```
