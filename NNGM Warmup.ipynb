```python
import random
import markovify
import nltk
import re
```

```python
novel = ''

asterisk = '***'

# Get raw text as string.
with open("nanowrimo/combined-allfinished-nanos-nofluff.txt") as f:
    text = f.read()
    
# Build the model.
text_model = markovify.Text(text)

# Make it obey sentence structure better. 
# The pos_tag is very slow but I can't get the alternative to work.
class POSifiedText(markovify.Text):
    def word_split(self, sentence):
        words = re.split(self.word_split_pattern, sentence)
        words = [ "::".join(tag) for tag in nltk.pos_tag(words) ]
        return words

    def word_join(self, words):
        sentence = " ".join(word.split("::")[0] for word in words)
        return sentence
    
# Print five randomly-generated sentences
for i in range(3600):
    
    novel += str(text_model.make_sentence(tries=20)) + ' '
    
    #if asterisk in novel:
     #   asterisk += '\n\n'
        
    r = random.randint(0,100)
    
    if (r<6):
        novel += '\n\n'
        
    #asterisk = asterisk.replace('***', '\n\n')
```
